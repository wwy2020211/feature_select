```python
import os
from collections import Counter
from tqdm import tqdm
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn import svm
from sklearn.model_selection import train_test_split, KFold
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,f1_score, confusion_matrix
from sklearn.feature_selection import SelectKBest,chi2,f_classif,SelectFromModel,RFE,VarianceThreshold,SelectPercentile,RFECV,f_regression
from sklearn.model_selection import learning_curve,cross_val_score
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier,RandomForestRegressor
from sklearn.neighbors import KNeighborsClassifier
```

```python
df = pd.read_csv('train_hf_data.csv')
x_train=np.array(df)[:,1:]
dfl= pd.read_csv('train_hf_label.csv')
y_train=np.array(dfl)[:,1]
y_train.astype(int)
pred=[]
#sgn(df,dfl)
#df=sigmoid(df)
ts = pd.read_csv('test_hf_data.csv')
x_test=np.array(ts)[:,1:]
tsl = pd.read_csv('test_hf_label.csv')
y_test=np.array(tsl)[:,1]
```

```python
def fetch(s1,s2,s3):
    #s1=[]
    df_feature=['age','anaemia','creatinine_phosphokinase','diabetes','ejection_fraction','high_blood_pressure','platelets','serum_creatinine','serum_sodium','sex','smoking','time']
    for i in range(len(s2)):
        if s2[i]==True:
            s1.append(np.asarray(s3[:,i]))
```

```python
classify=[RandomForestClassifier(n_estimators=50, max_depth=5, random_state=2),KNeighborsClassifier(),svm.LinearSVC()]
model_list=[SelectKBest(f_classif,4),RFE(estimator=svm.LinearSVC(),n_features_to_select=2),VarianceThreshold(1),RFECV(estimator=RandomForestRegressor(), step=0.1),SelectKBest(f_regression, k=5)]
model_name=['f_classif','svm-RFE','VarianceThreshold','FR-RFECV','f_regression']
classify_name=['随机森林','knn','svm']
k=0
#特征提取前的算法评价
t=0
for me in classify:
    print('分类器{}\n'.format(classify_name[t]))
    t=t+1
    tr=me.fit(x_train,y_train)
    y_pred = me.predict(x_train)
    y_pred_test = me.predict(x_test)
    acc_model = accuracy_score(y_train, y_pred)
    print('训练集准确度:{}'.format(acc_model))
    #print(y_pred_randomCLF_test)
    acc_model = accuracy_score(y_test, y_pred_test)
    print('测试集准确度:{}'.format(acc_model))
    cro_acc= sum(cross_val_score(me,x_train,y_train))/5
    print('交叉验证准确度：{}\n'.format(cro_acc))
#特征提取后的算法评价
for m in model_list:
    x_pre=m.fit_transform(x_train,y_train)
    x_pre.shape
    judge=m.get_support()
    print("选中特征有")
    print(judge)
    new_train=[]
    new_test=[]
    fetch(new_train,judge,x_train)
    new_train=(np.asarray(new_train)).T
    fetch(new_test,judge,x_test)
    new_test=(np.asarray(new_test)).T
    print('提取算法{}\n'.format(model_name[k]))
    k=k+1
    i=0
    for cla in classify:
        print('分类器{}\n'.format(classify_name[i]))
        i=i+1
        tr=cla.fit(new_train,y_train)
        y_pred = cla.predict(new_train)
        y_pred_test = cla.predict(new_test)
        acc_model = accuracy_score(y_train, y_pred)
        print('训练集准确度:{}'.format(acc_model))
        #print(y_pred_randomCLF_test)
        acc_model = accuracy_score(y_test, y_pred_test)
        print('测试集准确度:{}\n'.format(acc_model))
        cro_acc= sum(cross_val_score(cla,new_train,y_train))/5
        print('交叉验证准确度：{}\n'.format(cro_acc))
```
